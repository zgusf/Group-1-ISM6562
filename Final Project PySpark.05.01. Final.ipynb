```
df=df.dropna()
```

```
from pyspark.ml.regression import LinearRegression
from pyspark.ml.feature import VectorAssembler,StringIndexer
from pyspark.ml import Pipeline
```

```
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
```

```
df.columns
```

```
dfAssemblerFeature= VectorAssembler(
                inputCols=['bedrooms','bathrooms','sqft_living','sqft_lot','floors','waterfront','view','condition',
                           'grade','sqft_above','sqft_basement','yr_built','yr_renovated','zipcode','lat','long',
                           'sqft_living15','sqft_lot15'],
                outputCol="features")
    
```

```
print((df.count(),len(df.columns)))
```

```
train_data1,test_data1=df.randomSplit([0.7,0.3],23)
```

```
print((train_data1.count(),len(train_data1.columns)))
```

```
print((test_data1.count(),len(test_data1.columns)))
```

```
from pyspark.ml.regression import LinearRegression

lr = LinearRegression(labelCol='price')
```

```
# Chain indexer and tree in a Pipeline
pipeline_lr = Pipeline(stages=[dfAssemblerFeature, lr])
```

```
# Train model.  This also runs the assembler.
model_lr = pipeline_lr.fit(train_data1)
```

```
# Make predictions.
predictions_lr = model_lr.transform(test_data1)
```

```
from pyspark.ml.evaluation import RegressionEvaluator
# Select (prediction, true label) and compute test error
evaluator_lr = RegressionEvaluator(
    labelCol="price", predictionCol="prediction", metricName="rmse")
rmse_lr = evaluator_lr.evaluate(predictions_lr)
```

```
print("Root Mean Squared Error (RMSE) on test data = %g" % rmse_lr)
```

```
from pyspark.ml.regression import DecisionTreeRegressor
from pyspark.ml import Pipeline
```

```
# Train a DecisionTree model.
dt = DecisionTreeRegressor(labelCol='price')
```

```
# Chain indexer and tree in a Pipeline
pipeline_dt = Pipeline(stages=[dfAssemblerFeature, dt])
```

```
# Train model.  This also runs the assembler.
model_dt = pipeline_dt.fit(train_data1)
```

```
# Make predictions.
predictions_dt = model_dt.transform(test_data1)
```

```

print("Root Mean Squared Error (RMSE) on test data = %g" % rmse_dt)
```
